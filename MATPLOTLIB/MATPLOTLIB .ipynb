```python
# Import python library (matplotlib)
import matplotlib.pyplot as plt
```

* change parameter in graph

1] color - in default all color are changing (line & points)

2] marker - point (you can change style of points means you use symbol as you wish)
            (. , o,v,^,<,>,1,2,3,4,8,s,P,p,*,h,H,+,x)
            
3] linestyle(ls) - 1-solid line style (-)
                 2-dashed line styl  (--)                                            3-dash-dot line style (-.)        
                 4-dotted line style    (:)
                 
4] linewidth(lw) -To give linewidth (as you wish)

5] markersize(ms) -To give markersize (as you wish)

6] mfc - marker face color

7] mec - marker edge color

8] plt. xlable -  to give name of x axis (loc-'right','center','left')

9] plt.ylable -  to give name of y axis(loc- 'top','center','bottem')

10] plt.title - to give name of graph 

11] plt grid()- background lines :
    if you want only x axis then -plt.grid(axis=x)
    if you want only y axis then -plt.grid(axis=y)
    in default get both axis

```python
# if you dont want points connecting line
y=[5,9,6,9,12]
plt.plot(y,"*",ms=18)
plt.show()
```

```python
x =[2,3,4,5,6,7,8,9]
y =[7,5,8,3,9,7,5,8]
plt.plot(x,y,marker="*",ms=18,mfc="gold",mec="r",linestyle="-.",color="indigo")
plt.xlabel("Roll no",c="b")
plt.ylabel("marks ",c="b")
plt.title("class 5")
plt.grid(color='r',ls='-')
plt.show()
```

***** scatter graph *****

```python
import matplotlib.pyplot as plt
```

```python
# plot singal column scatter graph
age=[2,3,4,5,6,7,8,9]
speed=[12,23,36,65,54,64,65,98]
plt.xlabel("Age",fontsize="10")
plt.ylabel("Speed ",fontsize="10")
plt.title("Car",fontsize="15")
plt.scatter(age,speed, marker="*",color='b')
plt.show()
```

```python
# plot multiple column scatter graph

age=[2,3,4,5,6,7,8,9]
speed=[12,23,36,65,54,64,65,98]
speed1=[5,15,25,35,65,64,69,98]

plt.scatter(age,speed, marker="*",color="r")
plt.scatter(age,speed1,marker="<",color="g")
plt.xlabel("Age",fontsize="10")
plt.ylabel("Speed ",fontsize="10")
plt.title("Car",fontsize="15")
plt.legend(['speed','speed1'],loc="lower right",facecolor="pink")
plt.show()
```

***** Bar Graph *****

```python
import matplotlib.pyplot as plt
```

* If have singal singal values then use bar diagram

```python
# Vertical bar diagram

x=("pallavi","anjali","anuja","rucha")
y=(69,89,57,69)
plt.bar(x,y,width=0.5,ec="r")
plt.xlabel("name",color="r")
plt.ylabel("mark",color="r")
plt.show()
```

```python
# Horizontal bar diagram

x=("pallavi","anjali","anuja","rucha")
y=(69,89,57,60)
plt.barh(x,y,height=0.5,color="pink",ec="gold")
plt.xlabel("marks",color="b")
plt.ylabel("student",color="b")
plt.title("class-5",color="r")
plt.show()
```

***** PIE GRAPH *****

```python
y=[40,10,20,30]
x=["A","B","C","D"]
plt.pie(y,labels=x)
plt.show()
```

```python
# Focus on one part
y=[40,10,20,30]
x=["A","B","C","D"]
d=['MediumVioletRed',"DarkSalmon","r","LawnGreen"]
plt.pie(y,labels=x,colors=d,explode = [0.0,0.2,0.0,0.0])
plt.show()
```

```python
# Change angle
y=[40,10,20,30]
x=["A","B","C","D"]
d=['MediumVioletRed',"DarkSalmon","r","LawnGreen"]
plt.pie(y,labels=x,colors=d,explode = [0.0,0.2,0.0,0.0],startangle=180)
plt.show()
```

```python
# if you want shadow
y=[40,10,20,30]
x=["A","B","C","D"]
d=['MediumVioletRed',"DarkSalmon","r","LawnGreen"]
plt.pie(y,labels=x,colors=d,explode = [0.0,0.4,0.0,0.0],shadow=True)
plt.show()
```
