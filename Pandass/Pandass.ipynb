```python
# To Read csv file
df=pd.read_csv("D:\DATA SCIENCE\PANDAS\PRACTIES\dirtydata.csv.xls")
df.head()
```

```python
# To check information in dataframe & size of dataframe(memory) 
df.info()
```

```python
# To check blank values  
df.isnull().head()
```

```python
# To check blank values & count this values
df.isnull().sum()
```

```python
# Remove blank / null values
df.dropna().head()
```

```python
# Remove column wise null value (remove row wise)
df.dropna(subset=['Pulse','Date']).head()
```

```python
# To remove column in dataframe (remove column wise)
df.drop(['Maxpulse'],axis=1,inplace=True)
df
```

```python
# To add new column in dataframe
df['current year'] = 2022
df.head()
```

```python
# To describe int,flot type columns( min, max ,..)
df.describe()
```

```python
# To replace blank value 
df. fillna(100)
```

```python
# To replace perticular column wise
df['Pulse'].fillna(11111)
```

```python
# To find mean of perticular column
df['Calories'].mean()
```

```python
# To replace null value with mean value
cal_mean = df['Calories'].mean()
```

```python
df['Calories'].fillna(cal_mean)
```

```python
# To find median of perticular column
df['Calories'].median()
```

```python
# To replace null value with median value
cal_median = df['Calories'].median()
df['Calories'].fillna(cal_median)
```

```python
# To find mode of perticular column
df['Calories'].mode()
```

```python
# To replace null value with mode value
cal_mode = df['Calories'].mode()
df['Calories'].fillna(cal_mode)
```
